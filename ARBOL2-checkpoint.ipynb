```python
# Paso 0: Importar libreras necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.feature_selection import SelectKBest, f_classif

# Paso 1: Cargar datos
df = pd.read_csv("winequality-red.csv", sep=';')

# Paso 2: Crear variable objetivo binaria (vino bueno: calidad >= 7)
df['good_quality'] = (df['quality'] >= 7).astype(int)
df = df.drop('quality', axis=1)

# Paso 3: EDA rpido (opcional)
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlacin entre variables")
plt.show()

# Paso 4: Separar variables predictoras (X) y objetivo (y)
X = df.drop('good_quality', axis=1)
y = df['good_quality']

# Paso 5: Seleccin de caractersticas
selector = SelectKBest(score_func=f_classif, k=8)
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("Caractersticas seleccionadas:", list(selected_features))

# Paso 6: Dividir en Train y Test
X_train, X_test, y_train, y_test = train_test_split(
    X[selected_features], y, test_size=0.3, random_state=42, stratify=y)

# Paso 7: Entrenar el modelo (rbol de Decisin)
tree = DecisionTreeClassifier(max_depth=5, min_samples_split=10, random_state=42)
tree.fit(X_train, y_train)

# Paso 8: Evaluar el modelo
y_pred = tree.predict(X_test)
print("Reporte de clasificacin:\n", classification_report(y_test, y_pred))

# Matriz de confusin
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Prediccin")
plt.ylabel("Real")
plt.title("Matriz de confusin")
plt.show()

# Paso 9: Visualizar el rbol de decisin
plt.figure(figsize=(20, 10))
plot_tree(tree, feature_names=selected_features, class_names=['Malo', 'Bueno'], filled=True)
plt.title("rbol de decisin para calidad del vino")
plt.show()
```
